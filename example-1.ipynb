```python
import pandas as pd
```

```python
housing = pd.read_csv('cal_housing_clean.csv')
```

```python
housing.head()
```

```python
y_val = housing['medianHouseValue']
```

```python
x_data = housing.drop('medianHouseValue', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_val, test_size =0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = pd.DataFrame(data = scaler.transform(x_train),
                      columns=x_train.columns,
                      index =x_train.index)
```

```python
x_test = pd.DataFrame(data = scaler.transform(x_test),
                      columns=x_test.columns,
                      index =x_test.index)
```

```python
scaler.transform(x_train)
```

```python
housing.columns
```

```python
import tensorflow as tf
```

```python
age = tf.feature_column.numeric_column('housingMedianAge')
rooms = tf.feature_column.numeric_column('totalRooms')
beadrooms  = tf.feature_column.numeric_column('totalBedrooms')
pop  = tf.feature_column.numeric_column('population')
households = tf.feature_column.numeric_column('households')
income = tf.feature_column.numeric_column('medianIncome')
```

```python
feat_cols = [age, rooms, beadrooms, pop, households, income]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x= x_train,
                                                y = y_train,
                                                batch_size = 10,
                                                num_epochs =1000,
                                                shuffle = True)
```

```python
model = tf.estimator.DNNRegressor(hidden_units=[6,6,6], feature_columns=feat_cols)
```

```python
model.train(input_fn=input_func, steps=2000)
```

```python
predic_input_func = tf.estimator.inputs.pandas_input_fn(x= x_test,
                                                       batch_size =10,
                                                       num_epochs=1,
                                                       shuffle = False)
```

```python
pred_gen = model.predict(predic_input_func)
```

```python
predictions = list(pred_gen)
```

```python
predictions
```

```python
final_preds = []
for pred in predictions:
    final_preds.append(pred['predictions'])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, final_preds)**0.5
```
